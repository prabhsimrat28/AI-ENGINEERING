```python
#INHERTITANVE
class Animal:
    def speak(self):
        print("Animal")

class Dog(Animal):
    def bark(self):
        print("BArk")

d=Dog()
d.bark()
d.speak()
```

```python
#METHOD OVERRIDING IN INHERITANCE
class Animal:
    def speak(self):
        print("Animal")

class Dog(Animal):
    def speak(self):
        print("DOG")

d=Dog()
d.speak()
```

```python
#PASSING PARAMETERS TO BASE CLASS
class Animal:

    def __init__(self, name):
        self.name = name


class Dog(Animal):

    def __init__(self, name, breed):

        super().__init__(name)

        self.breed = breed


d = Dog("Tommy", "Labrador")

print(d.name)
print(d.breed)
```

```python
#CLASS METHODS ARE USED TO CHANGE THE CLASS VARIABLES THAT ARE COMMON FOR ALL OBJECTS OF THE CLASS
class Student:

    school = "DAV"

    @classmethod
    def change_school(cls, new):

        cls.school = new
```

```python
#PROPERTIES LET THE METHODS BEHAVE AS VARIABLES
class Circle:

    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        return 3.14 * self.radius * self.radius

c = Circle(5)

print(c.area)
```

```python
class Engine:

    def start(self):
        print("Started")


class Car:

    def __init__(self):

        self.engine = Engine()


c = Car()

c.engine.start()
```

```python
class CPU:

    def __init__(self, model):
        self.model = model


class Laptop:

    def __init__(self, brand, cpu_model):

        self.brand = brand
        self.cpu = CPU(cpu_model)


lap = Laptop("ASUS", "Intel i7")

print(lap.brand)
print(lap.cpu.model)
```
