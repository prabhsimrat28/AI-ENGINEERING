```python
#Count Vowels
s1="aaeioup"
count=0
for c in s1:
    if c in {"a","e","i","o","u"}:
        count=count+1

print(count)
```

```python
def fact(n):
    fac=1
    for i in range (1,n+1):
        fac=fac*i
    return fac

print(fact(5))
```

```python
#remove odd elements from a list

def fun(lst):
    ans=[]
    for i in lst:
        if i%2==0:
            ans.append(i)
    return ans

a=[2,5,8,3,1,6,0]
print(fun(a))
```

```python
def prime(a):
    for x in range (2,int(a/2)+1):
        if a%x==0:
            return 0
    return 1

print(prime(4))
```
