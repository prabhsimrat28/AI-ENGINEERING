```python
import torch

a = torch.tensor([2.0,3.0], requires_grad=True)
b= torch.tensor([6.0,4.0],requires_grad=True)

f=3*a**3 - b**2
print(f)
```

```python

f.backward(gradient=torch.tensor([1,1]))
print(a.grad)
print(b.grad)
```

```python

x=torch.tensor(2.0,requires_grad=True)
y=x**2
y.backward()
print(x.grad)

y=x**2
y.backward()
print(x.grad)
```
